Load Dataset SVHN_single_grey1.h5

```
import h5py
import numpy as np

SVHN=h5py.File('/content/drive/My Drive/Colab Notebooks/SVHN_single_grey1.h5','r')

#Load training and Test data

x_train=SVHN['X_train'][:]
y_train=SVHN['y_train'][:]
x_test=SVHN['X_test'][:]
y_test=SVHN['y_test'][:]
x_val = SVHN['X_val'][:]
y_val = SVHN['y_val'][:]

#Close File
SVHN.close()
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
#Reshape Data for KNN
x_train1=x_train.reshape(x_train.shape[0],32*32)
x_test1=x_test.reshape(x_test.shape[0],32*32)
```

```
#Applyig KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

knn= KNeighborsClassifier()
knn.fit(x_train1,y_train)
```

```
#Prediction
y_pred = knn.predict(x_test1)
```

```
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```
#Using Neural networks

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
y_train=keras.utils.to_categorical(y_train,10)
y_test=keras.utils.to_categorical(y_test,10)
y_val=keras.utils.to_categorical(y_val,10)
```

```
x_train2=x_train.reshape(x_train.shape[0],32,32,1).astype('float32')
x_test2=x_test.reshape(x_test.shape[0],32,32,1).astype('float32')
x_val2=x_val.reshape(x_val.shape[0],32,32,1).astype('float32')
```

```
#Normalizing the data
x_train2/=255
x_test2/=255
x_val2/=255
```

```
#Building Model
model=Sequential()
model.add(keras.layers.Flatten(input_shape=(32,32,1)))
model.add(Dense(200,activation ='relu'))
model.add(Dense(100,activation ='relu'))
model.add(keras.layers.Dropout(0.25))
model.add(Dense(10,activation='softmax'))
```

```
#Compile
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```
model.fit(x_train2,y_train, batch_size=64, epochs=10, 
              validation_data=(x_test2, y_test))
```

```
#Adding Batch normalizationa

model1=Sequential()
model1.add(keras.layers.Flatten(input_shape=(32,32,1)))
model1.add(Dense(200,activation ='relu'))
model1.add(Dense(100,activation ='relu'))
model1.add(Dense(50,activation ='relu'))
model1.add(keras.layers.normalization.BatchNormalization())
model1.add(Dense(10,activation='softmax'))
```

```
#Compile
model1.compile(loss='categorical_crossentropy',optimizer ='adam', metrics=['accuracy'])
```

```
model1.fit(x_train2,y_train, batch_size=64, epochs=10, 
              validation_data=(x_test2, y_test))
```

```
from sklearn.metrics import confusion_matrix
```

```
y_pre=model1.predict(x_val2)
```

```
y_pre1=model1.predict(x_test2)
```

```
target_result=np.argmax(y_pre,axis=1)
test_result=np.argmax(y_pre1,axis=1)
```

```
#validation report
print(classification_report(y_val,target_result))
print(confusion_matrix(y_val,target_result))
```
