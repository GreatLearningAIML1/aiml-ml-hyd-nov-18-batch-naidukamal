```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

### Reading Dataset



```python
df = pd.read_csv('bank.csv')
```

```python
df.head()
```

## dropping the columns RowNumber,customerID,surname as these are unique identifiers of the customer

```python

df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
df.head()
```

## Applying label encoding for columns Geography and Gender

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
le=LabelEncoder()
```

```python
df['Geography']=le.fit_transform(df['Geography'])
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df.head()
```

### Onehot encoding Column 'geography'


```python
df=pd.get_dummies(df,columns=['Geography'])
df.head()
```

### split features and target

```python
x=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

### Normalise the data

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)
```

### Load tensorflow

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# create first model
model = Sequential()
#Adding 3 dense layers
model.add(Dense(12, input_dim=12,init='uniform',activation ='relu'))
model.add(Dense(8,init='uniform',activation ='relu'))
model.add(Dense(1,init='uniform',activation ='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X_train, y_train, nb_epoch=20, batch_size=10,  verbose=2)
```

```python
#predecting the results
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(y_test, y_pred)
score
```
