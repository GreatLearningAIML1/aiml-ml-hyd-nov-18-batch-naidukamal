## Applying PCA to reduce the dimentions

```python
#Finding the components covering more than 95% of variance
```

```python
from sklearn.decomposition import PCA
from scipy.stats import zscore
```

```python
X=vehicle.drop('class',axis=1)
```

```python
x_z = X.apply(zscore)

array = x_z.values
```

```python
cov_matrix = np.cov(x_z.T)
print('Covariance Matrix \n%s', cov_matrix)
```

```python
# Step 2- Get eigen values and eigen vector
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print('Eigen Vectors \n%s', eigenvectors)
print('\n Eigen Values \n%s', eigenvalues)
```

```python
tot = sum(eigenvalues)
var_explained = [(i / tot) for i in sorted(eigenvalues, reverse=True)] 
cum_var_exp = np.cumsum(var_explained) 
```

```python
plt.bar(range(1,19), var_explained, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1,19),cum_var_exp, where= 'mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc = 'best')
plt.show()
```

```python
#Considering 7 components as it covers more than 95 percent of varience
```

```python
pca = PCA(n_components=7)
pca.fit(x_z)
```

```python
X_pca = pca.transform(x_z)
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_
```
