# Q2 Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?
# Q3 What conclusion does the test (p-value) lead to?

```python
# Perform a 2 sample T test 

# The null hypothesis state that the two populations being tested have no statistically significant difference. The alternate hypothesis will state that there is one present.

# Null Hypothesis : The old scheme and new scheme produce same results.
# Alternative Hypothesis : The old scheme and new scheme are different and hence the scheme has an impact of Sales

```

```python
dfpro
```

```python
from scipy import stats
```

```python
Olddf = dfpro['Old Scheme (in thousands)']
Newdf = dfpro['New Scheme (in thousands)']
```

```python
ttest = stats.ttest = ttest_ind(Olddf, Newdf, equal_var=False)
ttest
```

```python
# p value > 0.05, hence Null hypothesis is accepted. There is no significant difference between old and new scheme.
```
